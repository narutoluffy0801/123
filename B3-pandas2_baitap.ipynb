Việc hợp nhất các cột có thể được thực hiện bằng phương thức concat, merge của pandas.

```python
# Đọc file dữ liệu 'FoodPrice_in_Turkey.csv'
import pandas as pd
df=pd.read_csv('FoodPrice_in_Turkey.csv')
df.head(10)
# Xóa các dòng có thuộc tính ProductID trùng nhau, 
# giữ lại bản ghi cuối cùng, giữ chỉ số ban đầu của các dòng
df=df.drop_duplicates(['ProductId'], keep='last')
# Xóa các dòng có thuộc tính ProductID trùng nhau, 
# giữ lại bản ghi cuối cùng, thiết lập lại chỉ số
df=df.drop_duplicates(['ProductId'],keep='last').reset_index(drop=True)
# Tách file chứa thông tin sản phẩm
df_pro = df.loc[:,['ProductId','ProductName','UmId','UmName']]
df_pro
# Tách file chứa thông tin giá
df_pri = df.loc[:,['ProductId','Place','Month','Year','Price']]
df_pri
# Tách file chứa thông tin giá với số dòng từ bản ghi 10 đến 20
df_pri10 = df.loc[10:20,['ProductId','Place','Month','Year','Price']]
df_pri10
```

2. Ghép các cột từ các file
Cách 1: Dùng phương thức merge ghép 2 dataFrame có cùng chung một cột thuộc tính:pd.merge(dataFram1,dataFram2,on=’Tên thuộc tính chung’).
Cách 2: Dùng phương thức concat cho phép ghép từ hai dataFrame trở lên lại với nhau:pd.concat([dataFram1,dataFram2,…], axis=1).

```python
df1=pd.merge(df_pro,df_pri, on='ProductId')
df1
df2=pd.concat([df_pro,df_pri], axis=1)
df2
df2=pd.concat([df_pro,df_pri,df_pri10], axis=1)
df2
```

Việc hợp nhất các dòng có thể được thực hiện bằng phương thức concat, append của pandas.
Cách 1: Dùng phương thức concat cho phép ghép từ hai dataFrame trở lên lại với nhau:
pd.concat([dataFram1,dataFram2,…], axis=0)
Cách 2: Dùng phương thức append cho phép ghép hai dataFrame:
DataFrame.append(other, ignore_index=False, verify_integrity=False, sort=None)

```python
import pandas as pd
df=pd.read_csv('FoodPrice_in_Turkey.csv')
print(df.head())
# Tách file 1 chứa 5000 bản ghi đầu tiên
df1 = df.loc[0:4999,:]
df1
# Tách file 2 chứa các bản ghi còn lại
df2 = df.loc[5000:7380,:]
df2
# Tách file 3 chứa thông tin giá với số dòng từ bản ghi 1000 đến 2000
df3 = df.loc[1000:2000,['ProductId','Place','Month','Year','Price']]
df3
df4=pd.concat([df1,df2], axis=0)
df4
df5=pd.concat([df1,df2,df3], axis=0)
df5
df6=df1.append(df2)
df6
df7=df1.append(df3)
df7
```

Tính giá trị lớn nhất và nhỏ nhất của GDP.
Với mỗi châu lục hãy tính tổng GDP; trung bình cộng GDP. Hợp nhất 2 bảng này thành một bảng duy nhất gồm 3 thông tin: Tên châu lục; Tổng GDP; TBC GDP.

```python
import pandas as pd
df=pd.read_csv('GDPlist.csv')
print(df.info())
print(df.head(10))
# Truy cập các dòng trong cột "GDP"
df_GDP = df.loc[:,'GDP (millions of US$)']
print(df_GDP.count())   # đếm số biến GDP
print(df_GDP.min())     # tính giá trị nhỏ nhất
print(df_GDP.max())     # tính giá trị lớn nhất 
df_total=df.groupby('Continent')['GDP (millions of US$)'].sum()
df_total.head(10)
df_DCA=df.groupby('Continent')['GDP (millions of US$)'].mean()
df_DCA.head(10)
df_new=pd.merge(df_total,df_DCA,on='Continent')
df_new.head(10)
```

Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia. 
Xây dựng bảng Pivot table, với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.
Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.
Xây dựng bảng Pivot table cho biết tổng số lượng hàng bán được của mỗi ngày.

```python
import pandas as pd
df=pd.read_csv('OnlineRetail.csv')
print(df.info())
print(df.head(10))
pivot1=df.pivot_table(values='Quantity',index='InvoiceNo',columns='Country',aggfunc ='sum')
pivot1.head()
pivot2=df.pivot_table(values='Quantity', index='StockCode',aggfunc = {min, max})
pivot2.head()
import numpy as np
pivot3=df.pivot_table(values=['Quantity','UnitPrice'], index='StockCode',aggfunc ={'Quantity': np.sum,'UnitPrice': np.mean})
pivot3.head()
#pivot4=df.pivot_table(values='Quantity', index='InvoiceDate',aggfunc ='sum')
pivot4=df.pivot_table(values='Quantity', columns='InvoiceDate',aggfunc ='sum')
pivot4.head()
```
