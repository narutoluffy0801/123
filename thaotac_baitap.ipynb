```python
import pandas as pd
df=pd.read_csv('OnlineRetail.csv')
print(df.shape)
print(df.dtypes)
df.head()
```

```python
df['date']=df['InvoiceDate'].str.split(' ').str[0]
df['time']=df['InvoiceDate'].str.split(' ').str[1]

df['month']=df['date'].str.split('/').str[1]
print(df.dtypes)
df.head()
```

```python
import numpy as np
conditions = [(df['month']=='3')|(df['month']=='2')|(df['month']=='1'),
              (df['month']=='4')|(df['month']=='5')|(df['month']=='6'),
              (df['month']=='7')|(df['month']=='8')|(df['month']=='9'),
              (df['month']=='10')|(df['month']=='11')|(df['month']=='12'),]
              
choices = ['4','3','2','1']
df['Previous'] = np.select(conditions, choices, default='NA')
df.head()

```

```python
df['Amount'] = df['Quantity'] * df['UnitPrice'] 
df.head()
```

Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 5% nếu là ‘France’ ngược lại là 0%.

```python
import numpy as np
conditions = [(df['Country']=='United Kingdom')&(df['Previous']=='4'),
              (df['Country']=='France')]
              
choices = [0.1,0.05]
df['Discount'] = np.select(conditions, choices, default=1)

row = df.iloc[203]
print(row)
df.head()
```
